```python
import sklearn
import sklearn.datasets
X = sklearn.datasets.load_breast_cancer(as_frame=True)["data"]
y = sklearn.datasets.load_breast_cancer(as_frame=True)["target"]

import sklearn.model_selection
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3, random_state=1)
print(X_train.shape, X_test.shape)
```

```python
train = X_train
train["target"] = y_train
test = X_test
test["target"] = y_test

TRAIN_CSV = "train.csv"
TEST_CSV = "test.csv"
train.to_csv(TRAIN_CSV, index=False)
test.to_csv(TEST_CSV, index=False)
```

```python
from autogluon import TabularPrediction as task
time_limits = 30
predictor = task.fit(train_data=task.Dataset(file_path=TRAIN_CSV), label="target", time_limits=time_limits, presets='best_quality')
```

```python
predictions = predictor.predict(task.Dataset(file_path=TEST_CSV))
print("Accuracy score ", sklearn.metrics.accuracy_score(y_test, predictions))
```
